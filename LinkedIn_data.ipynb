## Login into the LinkedIn Account

```python
user = driver.find_element(By.XPATH,'//input[@id="username"]')
# Put here username for login
user.send_keys("USER NAME")

password = driver.find_element(By.XPATH,'//input[@id="password"]')
# Put here password for login
password.send_keys("PASSWORD")

submit = driver.find_element(By.XPATH,"//button[@class='btn__primary--large from__button--floating']")
submit.click()
sleep(5)
```

## Search the job title and location

```python
search_bar = driver.find_element(By.XPATH,'//input[@autocomplete="organization-title"]')
search_bar.send_keys("Web Scraping")

location = driver.find_element(By.XPATH,'//input[@autocomplete="address-level2"]')
location.send_keys("India")

sub = driver.find_element(By.XPATH,'//*[@id="global-nav-search"]/div/div[2]/button[1]')
sub.click()
sleep(3)
```

## Scroll the job webpage

```python
for a in range(1,10):
    scrolls = driver.find_elements(By.XPATH,'//div[@data-view-name="job-card"]')
    for sc in scrolls:
        driver.execute_script("arguments[0].scrollIntoView();",sc)
        #sleep(5)
```

## Scrape the data of jobs

```python
jobs = []
companies = []
locat = []
for i in range(1,26):
    link = driver.find_element(By.XPATH,"/html/body/div[5]/div[3]/div[4]/div/div/main/div/div[2]/div[1]/div/ul/li[{}]/div/div/div[1]/div[1]/div[2]/div[1]/a".format(str(i)))
    link.click()
    Job = driver.find_element(By.XPATH,"//h1/a")
    jobs.append(Job.text)
    company = driver.find_element(By.XPATH,"(//a[@class='app-aware-link '])[2]")
    companies.append(company.text)
    loc = driver.find_element(By.XPATH,"//div[@class='job-details-jobs-unified-top-card__primary-description-without-tagline mb2']")
    locat.append(loc.text)
    #sleep(5)
```

## Create the dataframes of jobs data

```python
data = pd.DataFrame()
data["Job Title"] = jobs
data["Company Name"] = companies
data["Job Location"] = locat

data
```

## Data Exporting in the CSV format

```python
data.to_csv("Jobs data.csv")
```
